```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("===Zomato Sales Analysis=== \n")

print("Sales Data Shape:", sales_data.shape)

print("Sample Data for first three restaurant: \n", sales_data[:3])
```

```python
# Total sales per year of all restaurant
print("Yearly Total Of All Restaurant:",np.sum(sales_data, axis = 0))

yearlyTotal = np.sum(sales_data[:, 1:], axis = 0)
print("Yearly Total Of All Restaurant:", yearlyTotal)
```

```python
# Minimum Sales Per Restaurant
min_sales = np.min(sales_data[:, 1:], axis = 1)
print(f"Minimum Sales Per Restaurant: {min_sales}")
```

```python
# Maximum Sales Per Year
max_sales = np.max(sales_data[:, 1:], axis = 0)
print(f"Maximum Sales Per Year: {max_sales}")
```

```python
# Average Sales Per Restaurant
avg_sales = np.mean(sales_data[:, 1:], axis = 1)
print(f"Average Sales Per Restaurant: {avg_sales}")
```

```python
#Cumulative Sales
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print(f"Cumulative Sales: {cumsum}")
```

```python
#plotting
plt.figure(figsize = (10,6))
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average Cumulative Sales Across All Restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector Addition: ", vector1 + vector2)
print("Vector Multiplication: ", vector1 * vector2)
print("Vector Dot Product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Vector Angle:", angle)
```

```python
restaurant_types = np.array(["biryani", "chinese", "pizza", "burger", "cafe"])

vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper:", vectorized_upper(restaurant_types))
```

```python
# Monthly Average
monthly_avg = sales_data[:, 1:] / 12
print("Monthly Average Sales: ", monthly_avg)
```
