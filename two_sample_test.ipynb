### TWO SAMPLE TEST

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
group_A = [85,88,90,92,87,85,89,91,86,88]
group_B = [82,84,80,83,81,79,78,85,84,83]

significance_value = 0.05
```

```python
t_stats, p_value = stats.ttest_ind(group_A,group_B, equal_var=False)

print(f"T value is {t_stats}")
print(f"P value is {p_value}")
```

```python
if p_value <= significance_value:
    print("I reject the null hypothesis")
else:
    print("I fail to reject the null hypothesis")
```
