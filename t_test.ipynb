```python
import numpy as np 

# importing stats from scipy
from scipy import stats
```

```python
sample_data = [172,174,168,171,173,175,178,169,170]
population_mean = 170
significance_value = 0.05
```

```python
# finding sample mean
sample_mean = np.mean(sample_data)
print(f"Sample Mean is {sample_mean}")

# finding sample standard deviation
sample_standard_deviation = np.std(sample_data, ddof=1) # ddof is degree of freedom
print(f"Sample Standard Deviation is {sample_standard_deviation}")

# finding sample size
sample_size = len(sample_data)
print(f"Sample Size is {sample_size}")
```

```python
# finding t-score
t_score = (sample_mean - population_mean) / (sample_standard_deviation / np.sqrt(sample_size))
print(f"T-Score is {t_score}")

# finding degree of freedom (df)
degree_of_freedom = sample_size - 1
print(f"Degree of freedom (df) is {degree_of_freedom}")
```

### for Two Tail T-test

```python
# finding critical values
t_table = stats.t.ppf(1-(significance_value/2), degree_of_freedom)
print(f"T-table is {t_table}")
print(f"So, for two tail t-test, Critical Values is {-t_table} to {t_table}")

# concluding T-test
if t_score <= -t_table:
    print("I reject null hypothesis")
elif t_score >= t_table:
    print("I reject null hypothesis")
else:
    print("I failed to reject null hypothesis")    
```

### for Left TailT-test

```python
# finding critical values
t_table = stats.t.ppf(significance_value, degree_of_freedom)
print(f"So, Critical Values is {t_table}")

# concluding T-test
if t_score <= -t_table:
    print("I reject null hypothesis")
else:
    print("I failed to reject null hypothesis")
```

### for Right Tail T-test

```python
# finding critical values
t_table = stats.t.ppf(1-significance_value, degree_of_freedom)
print(f"So, Critical Values is {t_table}")

# concluding T-test
if t_score >= -t_table:
    print("I reject null hypothesis")
else:
    print("I failed to reject null hypothesis")
```

# P-value using T-test

### for Two Tail T-test

```python
# Finding P-Value for two tail test of t-test
p_value = 2 * (1 - stats.norm.cdf(abs(t_score), degree_of_freedom))
print(f"P-Value is {p_value}")

# Concluding T-Test
if p_value <= significance_value:
    print("I reject the null hypothesis")
else:
    print("I failed to reject the null hypothesis or I accept the null hypothesis")
```

### for Left Tail T-test

```python
# Finding P-Value for left tail test of t-test
p_value = stats.norm.cdf(abs(t_score), degree_of_freedom)
print(f"P-Value is {p_value}")

# Concluding T-Test
if p_value <= significance_value:
    print("I reject the null hypothesis")
else:
    print("I failed to reject the null hypothesis or I accept the null hypothesis")
```

### for Right Tail T-test

```python
# Finding P-Value for Right tail test of t-test
p_value = 1 - stats.norm.cdf(abs(t_score), degree_of_freedom)
print(f"P-Value is {p_value}")

# Concluding T-Test
if p_value <= significance_value:
    print("I reject the null hypothesis")
else:
    print("I failed to reject the null hypothesis or I accept the null hypothesis")
```
