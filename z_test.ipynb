```python
import numpy as np

# importing normal distribution from scipy package inside scikit-learn
from scipy.stats import norm
```

```python
sample_data = [172,174,168,171,173,175,178,169,170]
population_mean = 170
population_std = 3
significance_value = 0.05
```

```python
# Finding Sample Mean
sample_mean = np.mean(sample_data)
print(f"Sample Mean is {sample_mean}")

# Finding Sample Size
sample_size = len(sample_data)
print(f"Sample Size is {sample_size}")
```

```python
# Finding Z-Score
z_score = (sample_mean - population_mean) / (population_std/np.sqrt(sample_size))
print(f"Z-Score is {z_score}")
```

```python
# Finding P-Value
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(f"P-Value is {p_value}")
```

```python
# Concluding Z-Test
if p_value < significance_value:
    print("I will reject the null hypothesis")
else:
    print("I failed to reject the null hypothesis or I accept the null hypothesis")
```
