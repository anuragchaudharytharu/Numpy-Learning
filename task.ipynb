```python
# Create a NumPy array of numbers from 10 to 50.
arr1 = np.arange(10,50)
print(arr1)
```

```python
# Generate a 3×3 identity matrix.
idMatrix = np.eye(3,3)
print(idMatrix)
```

```python
# Create a NumPy array of 10 random integers between 1 and 100.
arr2 = np.random.randint(1,100,10)
print(arr2)
```

```python
# Find the maximum and minimum of an array.
maximumNumber = np.max(arr2)
minimumNumber = np.min(arr2)
print(f"maximum: {maximumNumber}")
print(f"minimum: {minimumNumber}")
```

```python
# Reshape a 1D array of 12 elements into a 3×4 matrix.
org_arr = np.random.randint(1,100,12)
reshapedArray = np.reshape(org_arr, (3,4))
print(f"reshapedArray: \n {reshapedArray}")
```

```python
# Given an array Given an array [1,2,3,4,5], compute the square of each element. compute the square of each element.
arr3 = np.array([1,2,3,4,5])
print(arr3)

squaredArray = arr3**2
print(squaredArray)
```

```python
# Create a 5×5 array of random numbers and find its mean.
arr4 = np.random.randint(5,50,(5,5))
print(arr4)

print(f"Mean of the array: {np.mean(arr4)}")
```

```python
# Find the index of the maximum element in [10, 20, 50, 40, 30].
arr5 = np.array([10, 20, 50, 40, 30])
print(f"array5: {arr5}")

maxNum = np.max(arr5)
print(f"Maximum Number in the array: {maxNum}")

index = 0
for i in arr5:
    if i == maxNum:
        break
    else:
        index += 1
print(f"Index of the Maximum Number: {index}")
```

```python
# Create two arrays and perform element-wise addition, subtraction, multiplication, and division.
first_Array = np.random.randint(1,100,(3,4))
second_Array = np.random.randint(1,100,(3,4))
print(f"first array:\n{first_Array}\n\nsecond array:\n{second_Array}\n")
```

```python
multiplication = np.multiply(first_Array, second_Array)
print(f"Multiplication:\n {multiplication}\n")

division = np.divide(first_Array, second_Array)
print(f"Division:\n {division}\n")

addition = np.add(first_Array, second_Array)
print(f"Addition:\n {addition}\n")

subtraction = np.subtract(first_Array, second_Array)
print(f"Subtraction:\n {subtraction}\n")
```
